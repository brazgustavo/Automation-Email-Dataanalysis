```python
#Importing necessary libraries 📚
import time
import pandas as pd
import win32com.client


#Update database queries that are linked to SAP HANA 🔄
#Creat the application
File = win32com.client.Dispatch("Excel.Application")


#This makes the app not visible 🙈
File.Visible = 0


#Open the first file 📂
Workbook = File.Workbooks.open(r"C:\Users\f1p1xq8\Desktop\Python\Análise de dados\Itens sem Qnote\QM10.xlsx")


#Update the data 🔄
Workbook.RefreshAll()


#Wait some seconds ⏰
time.sleep(20)


#Save 💾
Workbook.Save()


#Wait 1 second ⏰
time.sleep(1)


#Quit 🚪
File.Quit()


#This makes the app not visible 🙈
File.Visible = 0


#Open the second file 📂
Workbook = File.Workbooks.open(r"C:\Users\f1p1xq8\Desktop\Python\Análise de dados\Itens sem Qnote\MB52 Livres.xlsx")


#Update the data 🔄
Workbook.RefreshAll()


#Wait some seconds ⏰
time.sleep(20)


#Save 💾
Workbook.Save()


#Wait 1 second ⏰
time.sleep(1)


#Quit 🚪
File.Quit()


#ler tabela MB52 para itens bloqueados 📊
mb52 = pd.read_excel('C:\\Users\\f1p1xq8\\Desktop\\Python\\Análise de dados\\Itens sem Qnote\\MB52 Livres.xlsx', sheet_name=None)


#Ler QM10 📊
qm10 = pd.read_excel('C:\\Users\\f1p1xq8\\Desktop\\Python\\Análise de dados\\Itens sem Qnote\\QM10.xlsx', sheet_name=None)


resultado = []
for material in mb52['Material']:
    if material not in qm10['Material'].values:
        resultado.append(material)

        
print(f'Temos {len(resultado)} materiais sem Qnote Livres no i140, i800')
print(f'Os itens são {resultado}.')

#Send the email to the warehouse managers 📩


#Creat the integration between Python and outlook 📩
outlook = win32com.client.Dispatch('outlook.application')
time_i140_df = pd.read_excel('Enviar E-mails.xlsx')


#create a personalized email for each supplier 📩
for i, email in enumerate(time_i140_df['E-mail']):
    integrante = time_i140_df.loc[i, 'Integrante']
    
    mail = outlook.CreateItem(0)
    mail.To = email
    mail.Subject = 'Relatório de Itens livres no I140 e I800 sem Qnote.'
    mail.Body = '''
    Prezado {}, 
    Segue abaixo a lista  de Itens Livres no I140 e I800 sem Qnote, conforme solicitado.
    Qualquer dúvida estou à disposição.
    Temos {} materiais sem Qnote Livres no i140, i800.
    Os itens são {}
    Att.,
    '''.format(integrante,len(resultado),resultado)
    mail.Send()


        
```
